# Task 2 <a id='2'></a>

```python
query = ''' WITH a AS (SELECT reviews.book_id,
            COUNT(review_id),
            AVG(rating)
            FROM reviews
            JOIN ratings ON reviews.book_id = ratings.book_id
            GROUP BY reviews.book_id)
            
            SELECT books.title,
            a.count AS n_reviews,
            a.avg AS avg_rating
            FROM a
            LEFT JOIN books
            ON a.book_id = books.book_id
            ORDER BY n_reviews desc, avg_rating DESC
        '''

pd.io.sql.read_sql(query, con = engine)
```

```python
query = '''
        WITH a AS (
        SELECT DISTINCT reviews.book_id
        FROM reviews
        JOIN books
        ON reviews.book_id = books.book_id),
        
        b AS (
        SELECT books.book_id,
               books.author_id,
               books.title,
               books.num_pages,
               books.publication_date,
               books.publisher_id
        FROM a
        JOIN books
        ON a.book_id = books.book_id),
        
        c AS (
        SELECT book_id
        FROM b),
        
        d AS (SELECT *
              FROM books
              WHERE book_id NOT IN(SELECT book_id FROM c)),
        
        e AS (SELECT title,
        AVG(rating) as rating
        FROM d
        JOIN ratings ON d.book_id = ratings.book_id
        GROUP BY title)
        

        SELECT *
        FROM e
        ORDER BY rating DESC
        '''
        
pd.io.sql.read_sql(query, con = engine)
```

Those books weren't reviewed (`n_reviews` = 0)

# Task 3 <a id='3'></a>

```python
query = ''' SELECT count(publishers.publisher),
            publishers.publisher
            FROM books join
            publishers on books.publisher_id = publishers.publisher_id
            where books.num_pages > 50
            group by publishers.publisher
            order by count desc
            limit 1
        '''

pd.io.sql.read_sql(query, con = engine)
```

Penguin Books is the winner, with 42 published books (taking books with more than 50 pages). Average publisher published 3 books.

# Task 4 <a id='4'></a>

```python
query = ''' with a as (SELECT books.book_id,
            count(rating_id) as ratings_per_book,
            avg(rating)
            FROM books join
            authors on books.author_id = authors.author_id
            join ratings on books.book_id = ratings.book_id
            group by books.book_id),
            
            s as (select * from a
            where a.ratings_per_book > 49
            order by a.avg desc
            limit 1)
            
            select author,
            title,
            avg as avg_rating
            from s
            left join books on s.book_id = books.book_id
            left join authors on books.author_id = authors.author_id
        '''

pd.io.sql.read_sql(query, con = engine)
```

Among books with at least 50 ratings, J.K. Rowling is the author with the highest average book rating (4.4), for "Harry Potter and the Prisoner of Azkaban". Mary GrandPré is the illustrator. Average rating in the data is 4.

# Task 5 <a id='5'></a>

```python
query = ''' with a as (select username,
        count(distinct book_id) 
        from ratings
        group by username
        order by count desc),
        
        b as (select *
        from a
        where count > 50),
        
        c as (select b.username,
        count(review_id)
        from b
        join reviews
        on b.username = reviews.username
        group by b.username)
        
        select avg(count)
        from c
        '''

pd.io.sql.read_sql(query, con = engine)
```

The average user, among those who rated more than 50 books, submitted 24 reviews. The average reviews per user is 17.
